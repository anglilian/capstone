```python
import pandas as pd
import malaya
```

```python
# Filter for all Malay language articles
df = pd.read_excel("scraped_articles.xlsx", sheet_name="Sheet2")
bm_df = df[df.lang == "BM"]
strings = bm_df.article.tolist()
```

```python
# Use BERT topic modeller
bert = malaya.transformer.load(model='bert')
attention = malaya.topic_model.attention(strings, n_topics=10, vectorizer=bert)
bm_topics = attention.top_topics(10, top_n=10, return_df=True)
```

```python
# Change settings to show full pandas output
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
topics = attention.get_topics(10)
sent = [t[1] for t in topics]
pd.DataFrame(sent, columns=["Topic sentence"])
```
